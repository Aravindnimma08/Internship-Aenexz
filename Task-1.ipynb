```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("Largest is:", a)
elif b > c:
    print("Largest is:", b)
else:
    print("Largest is:", c)
```

```python
year = int(input("Enter year: "))

if year % 4 == 0:
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
ch = input("Enter a letter: ").lower()

if ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u':
    print("Vowel")
else:
    print("Consonant")
```

```python
n = int(input("Enter number: "))

if n % 5 == 0 and n % 11 == 0:
    print("Yes, divisible by both 5 and 11.")
else:
    print("No, not divisible by both 5 and 11.")
```

```python
n = int(input("How many numbers? "))
sum = 0
i = 1

while i <= n:
    sum = sum + i
    i = i + 1

print("Sum:", sum)
```

```python
n = int(input("Table of: "))
i = 1

while i <= 10:
    print(n, "x", i, "=", n * i)
    i = i + 1
```

```python
n = 4  

for row in range(1, n + 1):
    print(" " * (n - row) + "*" * (2 * row - 1))
```
