```python
import math

def circle_area(radius):
    area = math.pi * radius * radius
    return area

r = float(input("Enter radius: "))
print("Area =", circle_area(r))
```

```python
def largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= c:
        return b
    else:
        return c
      
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))
print("Largest =", largest(x, y, z))
```

```python
def my_len(s):
    count = 0
    for char in s:
        count += 1
    return count
    
text = input("Enter a string: ")
print("Length =", my_len(text))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
    
num = int(input("Enter a number: "))
if is_prime(num):
    print(num, "is prime.")
else:
    print(num, "is not prime.")
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

n = int(input("Enter n: "))
print("Fibonacci(", n, ") =", fibonacci(n))
```

```python
def reverse_str(s):
    if len(s) == 0:
        return s
    else:
        return reverse_str(s[1:]) + s[0]

text = input("Enter a string: ")
print("Reversed:", reverse_str(text))
```

```python
n = int(input("Enter n: "))
total = 0

for i in range(1, n + 1):
    total += i

print("Sum from 1 to", n, "=", total)
```

```python
numbers = [10, 3, 8, 1, 12, 7, 4]

count_even = 0
for num in numbers:
    if num % 2 == 0:
        count_even += 1

print("Even numbers count =", count_even)
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print("Prime numbers between 1 and 100:")
for num in range(1, 101):
    if is_prime(num):
        print(num, end=" ")
print()  
```
