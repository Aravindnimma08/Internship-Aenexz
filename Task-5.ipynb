```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['FarePerPerson'] = df['Fare'] / df['FamilySize']

print("\n Dataset with new features:")
print(df[['PassengerId', 'Age', 'Fare','SibSp','Parch', 'FamilySize', 'FarePerPerson']].head())
```

```python
analysis = df.groupby(['Pclass', 'Sex']).agg(
    AverageFare=('Fare', 'mean'),
    SurvivalRate=('Survived', 'mean'),
).reset_index()

print("\nGroupby analysis by Passenger Class and Gender:")
print(analysis)
```
